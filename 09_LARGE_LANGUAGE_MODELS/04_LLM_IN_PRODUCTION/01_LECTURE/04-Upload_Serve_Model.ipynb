```python
# --max_model_len should be set for Llama models 
# we haven't tested for other models. It corresponds to 
# the context window size 
# --tokenizer is the base model tokenizer 
# In the below example we used a fined-tuned Llama model 
# Therefore we are using the Llama-3.2-1B tokenizer

# In a terminal : 
vllm serve qxzjy/model_llama \
    --max_model_len=100000 \
    --tokenizer=meta-llama/Llama-3.2-3B \
    --chat-template=template.jinja
```
