## Monthly Income

```python
print(df.groupby("Gender")["MonthlyIncome"].median())
```

```python
px.box(df, x = "Gender", y = "MonthlyIncome")
```

It's mostly equal, even if women tends to earn a little bit more

```python
fig = px.scatter(df, x="Age", y="MonthlyIncome", trendline="ols")
fig.show()
```

```python
fig = px.box(df, x = "MonthlyIncome", y = "Age", orientation = "h")
fig.show()
```

In the figure above, we can see that young employees (age < 25) have significantly lower incomes than other employees. Also, it's interesting to notice that the dispersion of the incomes increases when the age increases. This means that income inequalities are worse for older employees!

```python
fig = px.box(df, x="Education", y="MonthlyIncome")
fig.show()
```

The income depends a bit on the education level, especially for highest level (5) who tend to earn more money than others.

```python
fig = px.box(df, x = "BusinessTravel", y = "MonthlyIncome")
fig.show()
```

There's no significant difference between employees who travel and employees who don't.

## Attrition

```python
df_attrition = df["Attrition"].value_counts().reset_index()
print(df_attrition)
```

```python
fig = px.pie(df_attrition, names="Attrition", values="count")
fig.show()
```

```python
fig = px.box(df, x="Attrition", y="JobSatisfaction")
fig.show()
```

We can see a clear difference in satisfaction between employees who quit their job and employees who don't. Only 25% of the employees quitting have a satisfaction > 3.

```python
fig = px.box(df, x="Attrition", y="JobSatisfaction", color="Gender")
fig.show()
```

Interestingly, there's a difference between women and men who quit their job. Indeed, men who quit their job rate their satisfaction more severely than women. This means that it will be harder to identify women willing to quit from their satisfaction level.

```python
df["Attrition"] = df["Attrition"].apply(lambda x : 1 if x == "Yes" else 0)
print(df)
```

```python
df_attrition_field = df.groupby("EducationField")["Attrition"].mean().reset_index()
print(df_attrition_field)
```

```python
fig = px.bar(df_attrition_field, x="EducationField", y="Attrition")
fig.show()
```

The field with highest attrition rate is Human Resources!
