### Least desirable attributes

```python

attr_transcode = {"attr1_1":"Attractive",
                  "sinc1_1":"Sincere",
                  "intel1_1":"Intelligent",
                  "fun1_1":"Fun",
                  "amb1_1":"Ambitious",
                  "shar1_1":"Shared Interests"}

graph_labels = {"attribute":"Attribute",
                "points":"Number of points (out of 100)"}

columns_names = ["attribute", "points"]

df_fem_not = pd.DataFrame(dataset[dataset["gender"]==0].loc[:,list(attr_transcode.keys())].mean().reset_index())
df_fem_not.columns = columns_names
df_fem_not["attribute"] = df_fem_not["attribute"].apply(lambda x : attr_transcode.get(x))
df_fem_not["min"] = df_fem_not["points"].apply(lambda x : True if x == df_fem_not["points"].min() else False)

display(df_fem_not.head(6))

fig_1 = px.bar(df_fem_not,
             x=df_fem_not["attribute"],
             y=df_fem_not["points"],
             title="Male attributes importance ",
             labels=graph_labels,
             category_orders={"attribute":list(attr_transcode.values())},
             color=df_fem_not["min"],
             color_discrete_map={True:px.colors.qualitative.Plotly[1], False:px.colors.qualitative.Plotly[0]}
             )

fig_1.update_layout(showlegend=False)

fig_1.show()
```

```python
df_mal_not = pd.DataFrame(dataset[dataset["gender"]==1].loc[:,list(attr_transcode.keys())].mean().reset_index())
df_mal_not.columns = columns_names
df_mal_not["attribute"] = df_mal_not["attribute"].apply(lambda x : attr_transcode.get(x))
df_mal_not["min"] = df_mal_not["points"].apply(lambda x : True if x == df_mal_not["points"].min() else False)

display(df_mal_not.head(6))

fig_2 = px.bar(df_mal_not,
             x=df_mal_not["attribute"],
             y=df_mal_not["points"],
             title="Male attributes importance ",
             labels=graph_labels,
             category_orders={"attribute":list(attr_transcode.values())},
             color=df_mal_not["min"],
             color_discrete_map={True:px.colors.qualitative.Plotly[1], False:px.colors.qualitative.Plotly[0]}
             )

fig_2.update_layout(showlegend=False)

fig_2.show()
```

### Trying correlation

```python
df_main_data = dataset.loc[:,:"shar1_1"]
display(df_main_data.shape)
display(df_main_data.head())
```

```python

df_main_data_corr = df_main_data.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
mask = ((df_main_data_corr > 0.5) | (df_main_data_corr < -0.5)) & (df_main_data_corr != 1)
df_main_data_corr = df_main_data_corr[mask]
df_main_data_corr = df_main_data_corr.dropna(how='all')
df_main_data_corr = df_main_data_corr.dropna(how='all', axis=1)
display(df_main_data_corr)
```

### Order of date vs match pourcentage

```python
df_order_match = df_main_data.groupby("order")[["match"]].mean().reset_index()
df_order_match["match_pourcentage"] = df_order_match["match"].apply(lambda x : x*100)
display(df_order_match)

graph_labels = {"order":"Number of date",
                "match_pourcentage":"Match pourcentage (%)"}

fig_3 = px.scatter(df_order_match,
                   x="order",
                   y="match_pourcentage",
                   title="Titre",
                   labels=graph_labels,
                   trendline="ols")
fig_3.show()
```
