```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/qxzjy/vscworkspace/dsfs-ft-34/eda_module/data/starbucks_reviews.csv")
print(df.shape)
print(df.dtypes)
df.head()
```

```python
#df["date"] = pd.dt.st
df["date"] = pd.to_datetime(df["date"], format="%B %d, %Y")
print(df.dtypes)
df.head()
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.day_of_week
```

```python
df["Review"] = df["Review"].str.lower()
df.head()
```

```python
df["count_starbucks"] = df["Review"].str.count("starbucks")
df["wordcount"] = df["Review"].str.split(" ").apply(lambda x: len(x))

df.head()
```
