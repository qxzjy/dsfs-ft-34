# Get loaded with Bitcoin

**API benchmark**

Coinbase:
- 10 000 requests per hour
- Does not require authentication for price data
- Only able to query one date and one crypto at a time

Coinlib:
- Between 60 and 180 requests per hour
- No historical data
- Need API key


CoinPaprika:
- 36 000 requests per hour
- Historical data limited on free plan
- No authentication required


BraveNewCoin:
- API key + authentication token valid for 24hours
- 200 000 requests per month


CoinGecko:
- No API key needed
- Rate limit of 50 calls/minute
- Possibility to get historical data and OHLC (Open High Low Close) prices

```python
import requests
import json
from datetime import datetime
```

```python
# url_base = 'https://api.coingecko.com/api/v3/'
# url_coins = 'coins/'
# url_markets = 'markets/'
# payload = {'vs_currency':'usd', 'order':'market_cap_desc', 'per_page':'250'}
# coins = requests.get('https://api.coingecko.com/api/v3/coins/markets', params = payload)

base_url = "https://api.coingecko.com/api/v3/"

crypto = requests.get(f"{base_url}coins/markets?vs_currency=usd&order=market_cap_desc&per_page=250")

print(crypto.json())
```

```python
top_crypto = crypto.json()[:10]
print(top_crypto)
```

```python
bitcoin_ohlc = requests.get(f"{base_url}coins/bitcoin/ohlc?vs_currency=usd&days=365")

dict_ohlc = bitcoin_ohlc.json()

for i in dict_ohlc:
     i[0] = datetime.fromtimestamp(i[0]/1000).strftime('%Y-%m-%d')

display(dict_ohlc)
```

We can't get data on a daily basis, because we request more tha a month of data.

Data granularity works as follow :
- 1 - 2 days: 30 minutes
- 3 - 30 days: 4 hours
- 31 days and beyond: 4 days

```python
top_crypto_id = [i["id"] for i in top_crypto]

# top_crypto_id = [top_crypto_id[0]]

top_crypto_ohlc = {}

for i in top_crypto_id:

    try:
        crypto_ohlc = requests.get(f"{base_url}coins/{i}/ohlc?vs_currency=usd&days=365")

        top_crypto_ohlc[i] = {}
        for y in crypto_ohlc.json():
            day = y[0]

            if type(day) != int:
                raise Exception
            
            day = datetime.fromtimestamp(int(str(day)[:-3])).strftime('%Y-%m-%d')

            top_crypto_ohlc[i][day] = y[1:]
    except:
        print(f"Issue with API, for coins/{i} => Status-code : {crypto_ohlc.status_code}")
```

```python
display(top_crypto_ohlc)
```

```python
with open('top_crypto_ohlc.json', 'w') as f:
    json.dump(top_crypto_ohlc, f)
```
